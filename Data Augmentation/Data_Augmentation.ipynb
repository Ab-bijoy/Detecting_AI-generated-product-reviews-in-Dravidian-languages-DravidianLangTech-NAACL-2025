```python
from googletrans import Translator

def back_translate(text, src_lang='ta', mid_lang='en'):
    translator = Translator()
    translated = translator.translate(text, src=src_lang, dest=mid_lang)
    translated_text = translated.text if translated and translated.text else text

    return translated_text

train_df_1['back_translated_DATA'] = train_df_1['DATA'].apply(back_translate)

print(train_df_1['back_translated_DATA'].head())


```

```python

import pandas as pd
import nlpaug.augmenter.word as naw

data = pd.read_csv('train_df_1_backtranslated.csv')

augmenter = naw.ContextualWordEmbsAug(model_path='bert-base-uncased', action="substitute")

def augment_text(text):
    try:
        augmented_segments = augmenter.augment(text)
        if augmented_segments:
            return augmented_segments[0]
        else:
            return text
    except Exception as e:
        print(f"Augmentation error for text: {text}, Error: {e}")
        return text
augmented_rows = []

progress_counter = 0

for index, row in data.iterrows():
    augmented_text = augment_text(row['back_translated_DATA'])
    augmented_rows.append([row['ID'], augmented_text, row['LABEL']])
    progress_counter += 1
    if progress_counter % 100 == 0:
        print(f"Processed {progress_counter} rows")

augmented_df = pd.DataFrame(augmented_rows, columns=['ID', 'back_translated_DATA', 'LABEL'])

final_df = pd.concat([data, augmented_df], ignore_index=True)

final_df.to_csv('final_augmented_data.csv', index=False)

print("Augmentation completed. Saved to 'final_augmented_data.csv'.")
```
